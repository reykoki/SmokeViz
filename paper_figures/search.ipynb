```python
density = 'Heavy'# 'Medium'
#density = "Heavy"#'Heavy' #'Light' 'Medium'
num_imgs = 3 #len(data_dict['test']['truth'])
print('num images', num_imgs)
num = 0
minimum_window = 3 # hr time window
#truth_fns = random.choices(data_dict['test']['truth'], k=num_imgs)
for truth_fn in truth_fns:
    if density in truth_fn:
        fn = truth_fn.split('/')[-1]
        dt = get_dt_from_fn(fn)
        lat, lon = get_center_lat_lon(fn)
        lat, lon = float(lat), float(lon)

        
        try:
            heavy_label_count = get_heavy_label_count(fn, data_loc="/scratch1/RDARCH/rda-ghpcs/Rey.Koki/Mie_SmokeViz/")
            
            time_window, row = find_large_window(fn, hr_diff=minimum_window)
            large_sza = find_large_sza(lat, lon, dt, res=1000, img_size=256)
            solar_close = close_to_sunrise_sunset(dt, lat, lon)
            if solar_close and large_sza and heavy_label_count > 1000 and time_window > timedelta(hours=minimum_window):
                print(fn)
                print('label count: ', heavy_label_count)
                plot_densities_from_processed_data(fn, data_loc="/scratch1/RDARCH/rda-ghpcs/Rey.Koki/Mie_SmokeViz/")
                
                num+=1
                if num >= num_imgs:
                    break
        except:
            pass
```
